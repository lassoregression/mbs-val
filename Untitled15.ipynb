```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Vasicek model parameters
kappa = 0.15  # Speed of mean reversion
theta = 0.05  # Long-term mean of the interest rate
sigma = 0.01  # Interest rate volatility
r0 = 0.03     # Initial interest rate
T = 30        # Time horizon in years
dt = 1 / 12   # Time step in years (monthly)
n_paths = 10000  # Number of paths to simulate

def generate_vasicek_paths(kappa, theta, sigma, r0, T, dt, n_paths):
    """Generate interest rate paths using the Vasicek model."""
    n_steps = int(T / dt)
    paths = np.zeros((n_paths, n_steps + 1))
    paths[:, 0] = r0
    for t in range(1, n_steps + 1):
        random_shock = np.random.normal(size=n_paths)
        paths[:, t] = (paths[:, t - 1] + 
                       kappa * (theta - paths[:, t - 1]) * dt + 
                       sigma * np.sqrt(dt) * random_shock)
    return paths

interest_rate_paths = generate_vasicek_paths(kappa, theta, sigma, r0, T, dt, n_paths)

def calculate_psa_smm(month):
    """Calculate the Single Monthly Mortality (SMM) rate based on the PSA model."""
    if month <= 30:
        cpr = 0.002 * (month / 30)
    else:
        cpr = 0.046  # Adjusted based on the screenshot data
    return 1 - (1 - cpr) ** (1 / 12)

# Mortgage-backed security details
coupon_rate = 0.045  # Annual coupon rate from the screenshot
initial_principal = 854347  # Initial principal from the screenshot
current_principal = 625677  # Current principal from the screenshot
remaining_months = 360 - 61  # 61 months have passed since the origination

def calculate_monthly_payment(principal, annual_rate, months):
    """Calculate the monthly mortgage payment using the amortization formula."""
    monthly_rate = annual_rate / 12
    return principal * (monthly_rate / (1 - (1 + monthly_rate) ** -months))

monthly_payment = calculate_monthly_payment(current_principal, coupon_rate, remaining_months)

def calculate_cash_flows(principal, coupon_rate, smm_rates, interest_rate_paths):
    """Calculate the MBS cash flows for each simulated interest rate path."""
    n_paths, n_months = interest_rate_paths.shape
    monthly_payment = calculate_monthly_payment(principal, coupon_rate, n_months - 1)
    cash_flows = np.zeros((n_paths, n_months))
    # Initialize principal_balance as float to avoid type conflicts in arithmetic operations
    principal_balance = np.full(n_paths, principal, dtype=float)

    for month in range(n_months):
        interest_payment = principal_balance * (coupon_rate / 12)
        principal_payment = np.minimum(monthly_payment - interest_payment, principal_balance)
        smm = smm_rates[month]
        prepayment = smm * (principal_balance - principal_payment)
        cash_flows[:, month] = interest_payment + principal_payment + prepayment
        principal_balance -= (principal_payment + prepayment)

    return cash_flows

# Pre-compute the SMM rates for the remaining months of the MBS
smm_rates = np.array([calculate_psa_smm(m) for m in range(1, remaining_months + 1)])
cash_flows = calculate_cash_flows(current_principal, coupon_rate, smm_rates, interest_rate_paths[:, :remaining_months])

def calculate_discounted_cash_flows(cash_flows, interest_rate_paths):
    """Calculate the present value of cash flows discounted by the simulated interest rate paths."""
    n_paths, n_months = cash_flows.shape
    discounted_cash_flows = np.zeros(n_paths)

    for path in range(n_paths):
        discount_factors = np.exp(-np.cumsum(interest_rate_paths[path, :]) * dt)
        discounted_cash_flows[path] = np.sum(cash_flows[path, :] * discount_factors)

    return discounted_cash_flows

discounted_cash_flows = calculate_discounted_cash_flows(cash_flows, interest_rate_paths[:, :remaining_months])
security_value = np.mean(discounted_cash_flows)

# Hypothetical Market Price (you should replace this with the actual market price if known)
market_price_percentage = 97.5  # Example: 97.5% of the par value
market_price = market_price_percentage / 100 * initial_principal  # Calculate market price based on initial principal

# Calculate the security value as the mean of discounted cash flows
security_value = np.mean(discounted_cash_flows)

# Calculate the market-based fair value (could be directly from a market price or adjusted in context)
fair_value = market_price  # Assuming the market price reflects the current fair value

# Print out the results
print(f"Calculated Security Value: ${security_value:,.2f}")
print(f"Market-based Fair Value: ${fair_value:,.2f}")

# Compare the values
if security_value > fair_value:
    print("The security appears to be undervalued in the market.")
elif security_value < fair_value:
    print("The security appears to be overvalued in the market.")
else:
    print("The security is fairly priced according to the model.")

# Plotting functions for visualization
def plot_interest_rate_paths(interest_rate_paths, n_display=10):
    plt.figure(figsize=(12, 8))
    for i in range(min(n_display, interest_rate_paths.shape[0])):
        plt.plot(interest_rate_paths[i, :], label=f'Path {i+1}')
    plt.title('Sample Interest Rate Paths')
    plt.xlabel('Months')
    plt.ylabel('Interest Rate')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_cash_flows(cash_flows, n_display=5):
    plt.figure(figsize=(12, 8))
    for i in range(min(n_display, cash_flows.shape[0])):
        plt.plot(cash_flows[i, :], label=f'Path {i+1}')
    plt.title('Sample Cash Flows')
    plt.xlabel('Months')
    plt.ylabel('Cash Flow ($)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the plotting functions
plot_interest_rate_paths(interest_rate_paths)
plot_cash_flows(cash_flows)
```

```python
def plot_discounted_cash_flow_distribution(discounted_cash_flows):
    """Plot the distribution of discounted cash flows."""
    plt.figure(figsize=(10, 6))
    plt.hist(discounted_cash_flows, bins=30, alpha=0.75, color='green')
    plt.title('Distribution of Discounted Cash Flows')
    plt.xlabel('Discounted Cash Flow ($)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

plot_discounted_cash_flow_distribution(discounted_cash_flows)
```
